## Var

### Local var
1. [a-z] or _
2. 局部变量是在方法内部声明的变量，它们只在声明它们的方法或代码块的作用域内可见。局部变量以小写字母或下划线开头。

### Instance var
1. @
2. 实例变量以 @ 符号开头，它们属于对象的特定实例，并且只能通过对象的方法访问。实例变量在对象之间是独立的，每个对象都有自己的一组实例变量。

### Class var
1. @@
2. 类变量以两个 @ 符号开头，它们属于整个类，并且在类的所有实例之间共享。类变量在整个类的范围内可见，并且可以通过类的任何实例或类方法访问。

### Global var
1. $
2. 全局变量以美元符号 $ 开头，可以在程序的任何地方访问。它们对整个程序都是可见的，即使是在不同的作用域或模块中也是如此。全局变量应该谨慎使用，因为它们可以在程序的任何地方修改，导致代码难以理解和维护

#### Evaluate a var
1. 字符串插值允许直接在字符串中向variable添加value

#### local var shadow
1. 如果代码内外有同名var，优先使用代码块内部的 __*python有相同特性*__

#### Instance Var & Class Var
1. 类似，前者只在特定instance of a obkect中，后者可以在所有实例中使用，通常用来更新状态
2. Instance Var is no need to initialize未经初始化会被赋nil, global var未初始化也赋nil
3. Class Var need to be initialized
